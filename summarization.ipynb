```python
# 2. Import libraries
from transformers import pipeline

```

```python
# 3. Load summarization pipeline
summarizer = pipeline("summarization")

```

```python
# 4. Your input text
text = """
Artificial intelligence (AI) refers to the simulation of human intelligence in machines
that are programmed to think and learn. AI is being used in a wide range of fields including
healthcare, finance, education, and more. It enables computers to perform tasks such as
decision-making, language translation, and speech recognition. As AI continues to evolve,
it is expected to bring about major changes in the way we live and work.
"""

```

```python
# 5. Generate summary
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

```

```python
# 6. Print result
print("Summary:", summary[0]['summary_text'])
```

```python
summary[0]['summary_text']
```

```python
# 3. Load summarization pipeline with the best model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
# 4. Your text to summarize
text = """
Artificial intelligence (AI) refers to the simulation of human intelligence in machines
that are programmed to think and learn. AI is being used in a wide range of fields including
healthcare, finance, education, and more. It enables computers to perform tasks such as
decision-making, language translation, and speech recognition. As AI continues to evolve,
it is expected to bring about major changes in the way we live and work.
"""

# 5. Summarize text
summary = summarizer(text, max_length=60, min_length=30, do_sample=False)

```
